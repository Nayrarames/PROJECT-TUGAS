Program Pengelolaan Nilai Mahasiswa

```python
nilai = []

# Input 10 nilai
for i in range(10):
    n = int(input(f"Masukkan nilai ke-{i+1}: "))
    nilai.append(n)

# Menghitung nilai tertinggi dan terendah
tertinggi = max(nilai)
terendah = min(nilai)

# Menghitung rata-rata
rata = sum(nilai) / len(nilai)

# Menghitung jumlah lulus
lulus = 0
for n in nilai:
    if n >= 60:
        lulus += 1

tidak_lulus = len(nilai) - lulus

# Menampilkan hasil
print("\n=== HASIL ANALISIS NILAI ===")
print("Nilai tertinggi :", tertinggi)
print("Nilai terendah  :", terendah)
print("Rata-rata nilai :", round(rata,2))
print("Jumlah lulus    :", lulus)
print("Tidak lulus     :", tidak_lulus)
```

Grafik Nilai Tertinggi dan Terendah

```python
print("\nGrafik Nilai")

print("Tertinggi :", "█" * (tertinggi // 2), tertinggi)
print("Terendah  :", "█" * (terendah // 2), terendah)
```

Grafik Kelulusan

```python
print("\nGrafik Kelulusan")

print("Lulus       :", "█" * lulus, lulus)
print("Tidak Lulus :", "█" * tidak_lulus, tidak_lulus)
```
